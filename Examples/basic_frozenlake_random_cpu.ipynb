<h2>Randomly playing games</h2>

```python
# import necessary modules
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output

# create the FrozenLake environment, DISABLE SLIPPERY
env = gym.make("FrozenLake-v1", is_slippery=False, render_mode="rgb_array")
```

<h3>Make CPU play the game RANDOMLY</h3>

```python
# reset the environment before playing the game
env.reset()

# 100 actions in a row
for _ in range(100):
    # get a random action between 0-3 (left, down, right, up)
    random_action = env.action_space.sample()

    # get the observation, reward etc. from the environment => MOVE THE ELF
    observation, reward, done, truncated, info = env.step(random_action)

    # clear the output below
    clear_output(wait=True)
    
    # render the next frame after user moved
    plt.imshow(env.render())
    plt.show()

    # if we either fall into a hole or find the GIFT
    if done:
        print(f"Game Over! Reward: {reward} - Resetting...")
        time.sleep(1.5)

        # before attempting to visualize the environment, a reset() a required
        env.reset()

        # clear the output below
        clear_output(wait=True)

        # show the current frame of the game by using Matplotlibd
        plt.imshow(env.render())
        plt.show()

    # between the steps, let's rest a while
    time.sleep(0.35)

# finally close the env when the CPU has had enough
env.close()
```
